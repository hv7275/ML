```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler

object_cols = df.select_dtypes(include=['object']).columns.to_list()
object_cols = object_cols + ['FastingBS']

ohe = OneHotEncoder(
    handle_unknown='ignore',
    drop='first',
    sparse_output=False
)

# Encode categorical features
encoded_df = pd.DataFrame(
    ohe.fit_transform(df[object_cols]),
    columns=ohe.get_feature_names_out(object_cols),
    index=df.index
)

# Drop original object columns
df = df.drop(object_cols, axis=1)

# Combine numeric + encoded categorical
final_df = pd.concat([df, encoded_df], axis=1)

scaled_df = final_df.copy()

numeric_cols = df.select_dtypes(exclude=['object']).columns.to_list()
numeric_cols.remove('HeartDisease')

scaler = StandardScaler()

scaled_df[numeric_cols] = scaler.fit_transform(scaled_df[numeric_cols])

scaled_df
```

```python
# Use only the requested models with manual preprocessing (no Pipelines)
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score

# Reload and preprocess explicitly (scale numerics + one-hot encode categoricals)
raw_df = pd.read_csv('../Data/heart.csv')
X = raw_df.drop('HeartDisease', axis=1)
y = raw_df['HeartDisease']

numeric_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
categorical_cols = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope', 'FastingBS']

scaler = StandardScaler()
ohe = OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False)

num_scaled = scaler.fit_transform(X[numeric_cols])
cat_encoded = ohe.fit_transform(X[categorical_cols])
X_processed = np.hstack([num_scaled, cat_encoded])

X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42, stratify=y
)

models = {
    "LogisticRegression": LogisticRegression(max_iter=500, n_jobs=-1),
    "KNN": KNeighborsClassifier(n_neighbors=11),
    "NaivBayes": GaussianNB(),
    "DecisionTree": DecisionTreeClassifier(max_depth=6, min_samples_leaf=2, random_state=42),
    "SupportVectorMachine": SVC(kernel='rbf', C=3, gamma='scale')
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

result = []
for name, model in models.items():
    cv_f1 = cross_val_score(model, X_processed, y, cv=cv, scoring='f1', n_jobs=-1)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    result.append({
        'Model': name,
        'CV_F1_mean': round(cv_f1.mean(), 3),
        'CV_F1_std': round(cv_f1.std(), 3),
        'Accuracy': round(acc, 3),
        'F1_Score': round(f1, 3)
    })

pd.DataFrame(result).sort_values(by='F1_Score', ascending=False)
```
